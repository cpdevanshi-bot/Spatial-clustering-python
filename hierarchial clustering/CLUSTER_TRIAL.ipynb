```python
from sklearn.preprocessing import StandardScaler

data = data.select_dtypes(include=['float64', 'int64'])

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
from scipy.cluster.hierarchy import linkage, cophenet
from scipy.spatial.distance import pdist

Z = linkage(data_scaled, method='ward')
c, _ = cophenet(Z, pdist(data_scaled))

print("Cophenetic Correlation Coefficient:", c)
```

```python
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import fcluster

for k in range(2, 10):
    labels = fcluster(Z, t=k, criterion='maxclust')
    score = silhouette_score(data_scaled, labels)
    print(f"K={k}, Score={score:.4f}")
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram

plt.figure(figsize=(10,5))
dendrogram(Z)
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load data
data = pd.read_excel(r"C:/Users/DEVANSHI/Desktop/COMBINED MATRIX_UPDATED.xlsx")

# Keep numeric columns
data = data.select_dtypes(include=['float64', 'int64'])

# Scale data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(data_scaled, method='ward')

plt.figure(figsize=(12,6))
dendrogram(Z)

plt.savefig(r"D:/MPLAN/thesis/dendrogram.pdf", bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(data_scaled, method='ward')

plt.figure(figsize=(16,7))

dendrogram(
    Z,
    no_labels=True,              # removes clutter
    truncate_mode='level',       # simplify tree
    p=5                          # show top levels
)

plt.savefig(r"D:/MPLAN/thesis/dendrogram_clean.pdf", bbox_inches='tight')
plt.show()
```
